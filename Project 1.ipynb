```python
gbadf = reldf[reldf['GBA'].notnull()]
```

```python
pricedf = gbadf[gbadf['PRICE'].notnull()]
```

```python
pricedf.describe()
```

```python
bins = [0, 1216, 1504, 1984, 15902]

group_names = ['Q1', 'Q2', 'Q3', 'Q4']

pricedf['size qrt'] = pd.cut(pricedf['GBA'], bins, labels=group_names)
pricedf.head()
```

```python
#Average price by quadrant
pricedf['PRICE'].groupby(pricedf['QUADRANT']).mean()
```

```python
#Average price by size
pricedf['PRICE'].groupby(pricedf['size qrt']).mean()
```

```python
pricedf.groupby(['QUADRANT','size qrt']).mean()
```

```python
pricedf.groupby(['QUADRANT', 'ASSESSMENT_NBHD']).mean()
```
